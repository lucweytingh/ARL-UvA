# Dataset overview usc adult

```python
from __future__ import division
import pandas as pd
import numpy as np
import json
import os,sys
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
pd.options.display.float_format = '{:,.2f}'.format
dataset_base_dir = '../datasets/uci_adult/'
train_file = 'train.csv'
test_file = 'test.csv'
```

```python
path_train = os.path.join(dataset_base_dir,train_file)
path_test = os.path.join(dataset_base_dir,test_file)

with open(path_train, "r") as train_fi:
    with open(path_test, "r") as test_fi: 
        train = pd.read_csv(train_fi, header=None)
        test = pd.read_csv(test_fi, header=None)
        df = train.append(test)
df
```

## General

```python
print('Size')
size = len(df)
print(size)

print('No. features')
nr_features = len(df.columns)
print(nr_features)

print('Protected features:')
print('Race, Sex')

print('Protected groups:')
print('{White, Black} x {Male, Female}')

print('Prediciton task:')
print('income ≥ 50k?')
```

## Distribution of sex

```python
print('Amount of Males:')
nr_males = (df[9] == 'Male').sum()
print(nr_males)

print('Percentage males:')
total_sex = len(df[9])
print(nr_males/total_sex)


print('Amount of Females:')
nr_females = (df[9] == 'Female').sum()
print(nr_females)

print('Percentage Females:')
print(nr_females/total_sex)
```

## Distribution of race

```python
print('Amount of White individuals:')
nr_white = (df[8] == 'White').sum()
print(nr_white)

print('Percentage White individuals:')
total_race = len(df[8])
print(nr_white/total_race)


print('Amount of Black individuals:')
nr_black = (df[8] == 'Black').sum()
print(nr_black)

print('Percentage Black individuals:')
print(nr_black/total_race)



print('Amount of Other individuals:')
nr_other = (df[8] == 'Other').sum() + (df[8] == 'Asian-Pac-Islander').sum() & (df[8] == 'Amer-Indian-Eskimo').sum()
print(nr_other)

print('Percentage Other individuals:')
print(nr_other/total_race)
```

## Distribution of income for race

```python
print('Amount of white individuals with income above 50K:')
nr_white_income = ((df[8] == 'White') & (df[14] == '>50K')).sum()
print(nr_white_income)

print('Percentage of white individuals with income above 50K:')
print(nr_white_income/nr_white)


print('Amount of white individuals with income below 50K:')
nr_white_n_income = ((df[8] == 'White') & (df[14] == '<=50K')).sum()
print(nr_white_n_income)

print('Percentage of white individuals with income below 50K:')
print(nr_white_n_income/nr_white)
```

```python
print('Amount of black individuals with income above 50K:')
nr_black_income = ((df[8] == 'Black') & (df[14] == '>50K')).sum()
print(nr_black_income)

print('Percentage of black individuals with income above 50K:')
print(nr_black_income/nr_black)


print('Amount of black individuals with income below 50K:')
nr_black_n_income = ((df[8] == 'Black') & (df[14] == '<=50K')).sum()
print(nr_black_n_income)

print('Percentage of black individuals with income below 50K:')
print(nr_black_n_income/nr_black)
```

## Distribution of income for sex

```python
print('Amount of males with income above 50K:')
nr_male_income = ((df[9] == 'Male') & (df[14] == '>50K')).sum()
print(nr_male_income)

print('Percentage of males with income above 50K:')
print(nr_male_income/nr_males)


print('Amount of males with income below 50K:')
nr_male_2_income = ((df[9] == 'Male') & (df[14] == '<=50K')).sum()
print(nr_male_2_income)

print('Percentage of males with income below 50K:')
print(nr_male_2_income/nr_males)
```

```python
print('Amount of females with income above 50K:')
nr_female_income = ((df[9] == 'Female') & (df[14] == '>50K')).sum()
print(nr_female_income)

print('Percentage of females with income above 50K:')
print(nr_female_income/nr_females)


print('Amount of females with income below 50K:')
nr_female_2_income = ((df[9] == 'Female') & (df[14] == '<=50K')).sum()
print(nr_female_2_income)

print('Percentage of females with income below 50K:')
print(nr_female_2_income/nr_females)
```
