# Dataset overview compas

```python
from __future__ import division
import pandas as pd
import numpy as np
import json
import os,sys
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
pd.options.display.float_format = '{:,.2f}'.format
dataset_base_dir = '../datasets/compas/'
train_file = 'train.csv'
test_file = 'test.csv'
```

```python
path_train = os.path.join(dataset_base_dir,train_file)
path_test = os.path.join(dataset_base_dir,test_file)

with open(path_train, "r") as train_fi:
    with open(path_test, "r") as test_fi: 
        train = pd.read_csv(train_fi, header=None)
        test = pd.read_csv(test_fi, header=None)
        df = train.append(test)
df
```

## Distribution of sex

```python
print('Amount of Males:')
nr_males = (df[8] == 'Male').sum()
print(nr_males)

print('Percentage males:')
total_sex = len(df[8])
print(nr_males/total_sex)


print('Amount of Females:')
nr_females = (df[8] == 'Female').sum()
print(nr_females)

print('Percentage Females:')
print(nr_females/total_sex)
```

## Distribution of race

```python
print('Amount of White individuals:')
nr_white = (df[9] == 'White').sum()
print(nr_white)

print('Percentage White individuals:')
total_race = len(df[9])
print(nr_white/total_race)


print('Amount of Black individuals:')
nr_black = (df[9] == 'Black').sum()
print(nr_black)

print('Percentage Black individuals:')
print(nr_black/total_race)



print('Amount of Other individuals:')
nr_other = (df[9] == 'Other').sum()
print(nr_other)

print('Percentage Other individuals:')
print(nr_other/total_race)
```

## Distribution of recid for race

```python
print('Amount of black people recid:')
nr_black_recid = ((df[9] == 'Black') & (df[10] == 'Yes')).sum()
print(nr_black_recid)

print('Percentage black recid:')
print(nr_black_recid/nr_black)


print('Amount of black people not recid:')
nr_black_n_recid = ((df[9] == 'Black') & (df[10] == 'No')).sum()
print(nr_black_n_recid)

print('Percentage black people not recid:')
print(nr_black_n_recid/nr_black)
```

```python
print('Amount of white people recid:')
nr_white_recid = ((df[9] == 'White') & (df[10] == 'Yes')).sum()
print(nr_white_recid)

print('Percentage white recid:')
print(nr_white_recid/nr_white)


print('Amount of white people not recid:')
nr_white_n_recid = ((df[9] == 'White') & (df[10] == 'No')).sum()
print(nr_white_n_recid)

print('Percentage white people not recid:')
print(nr_white_n_recid/nr_white)
```
